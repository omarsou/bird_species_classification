```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
%matplotlib inline
sns.set_context('poster')
sns.set_color_codes()
```

```
from matplotlib.font_manager import FontProperties
fontP = FontProperties()
fontP.set_size('xx-small')
```

```
order = ['inceptionv3', 'xception', 'inceptionresnet']
acc_orig_images_train = [0.9427, 0.9574, 0.9038]
acc_orig_images_val = [0.8738, 0.9029, 0.8932]
acc_crop_images_train = [0.8587, 0.86397, 0.8427]
acc_crop_images_val = [0.8804, 0.8587, 0.8478]
f1_orig_images_train = [0.9375, 0.9535, 0.8938]
f1_orig_images_val = [0.8777,  0.9044, 0.8777]
f1_crop_images_train = [0.8655, 0.8621, 0.8239]
f1_crop_images_val = [0.8801, 0.8682, 0.8536]
```

```
# Visual comparison
titles = ["Accuracy on Original Image", "Accuracy on Crop Images"]
y = [[acc_orig_images_train, acc_orig_images_val], [acc_crop_images_train , acc_crop_images_val]]
ncols = len(titles)
fig, axes = plt.subplots(1, ncols, figsize=(15, 5))
for idx, ax in enumerate(axes.flatten()):
    title = titles[idx]
    ax.plot(order, y[idx][0], label='Train', color='b', marker='.', markersize=18, linewidth=0)
    plt.xticks(rotation=9)
    ax.plot(order, y[idx][1], label='Validation', color='r', marker='.', markersize=18, linewidth=0)
    plt.xticks(rotation=9)
    ax.set_title(title)
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', prop=fontP)
fig.tight_layout()
```

```
# Visual comparison
titles = ["F1-score on Original Image", "F1-score on Crop Images"]
y = [[f1_orig_images_train, f1_orig_images_val], [f1_crop_images_train , f1_crop_images_val]]
ncols = len(titles)
fig, axes = plt.subplots(1, ncols, figsize=(15, 5))
for idx, ax in enumerate(axes.flatten()):
    title = titles[idx]
    ax.plot(order, y[idx][0], label='Train', color='b', marker='.', markersize=18, linewidth=0)
    plt.xticks(rotation=9)
    ax.plot(order, y[idx][1], label='Validation', color='r', marker='.', markersize=18, linewidth=0)
    plt.xticks(rotation=9)
    ax.set_title(title)
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', prop=fontP)
fig.tight_layout()
```
